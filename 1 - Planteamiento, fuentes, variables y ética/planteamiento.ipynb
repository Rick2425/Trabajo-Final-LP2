1. Planteamiento del problema 

En la actualidad, los medios digitales y plataformas de redes sociales generan grandes volúmenes de información en tiempo real, especialmente en el ámbito político y social. Esta información puede ser analizada mediante técnicas de web scraping y minería de datos para identificar patrones discursivos, encuadres informativos y reacciones de los usuarios. En este contexto, surge la necesidad de desarrollar un sistema de auditoría de fuentes que permita evaluar previamente si una página web es apta para scraping, si requiere el uso de una API o si presenta restricciones que impiden su extracción directa.

2. Justificación del proyecto
 
El presente proyecto es relevante porque permite analizar cómo los encuadres discursivos presentes en noticias políticas pueden influir o reflejarse en las reacciones de los usuarios en plataformas digitales.

Además, el uso de técnicas de web scraping y APIs permite estructurar grandes volúmenes de información no organizada, facilitando su análisis posterior mediante minería de datos.

Este enfoque es útil en el estudio de fenómenos como la desinformación, la polarización política y la construcción del discurso mediático en entornos digitales.

3. Fuentes de Información 
Las fuentes seleccionadas para el proyecto son: 

* La Republica 

Medio digital peruano con alta producción de noticias políticas nacionales.
Permite el acceso a contenido público en formato HTML, lo que lo hace adecuado para técnicas de scraping no intrusivo.

* The New York Times 

Medio internacional con cobertura política estructurada.
Debido a sus restricciones de acceso, se recomienda el uso de su API oficial para la extracción de datos.

* Youtube 

Plataforma de contenido audiovisual donde los usuarios generan comentarios públicos sobre temas políticos.
Por sus restricciones técnicas, la extracción de datos debe realizarse mediante la YouTube Data API.
